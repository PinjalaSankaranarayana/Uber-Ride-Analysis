```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
d=pd.read_csv("/content/ncr_ride_bookings.csv")
print(d)
```

```python
df=pd.DataFrame(d)
print(d)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df = df.rename(columns={'Date': 'Booking_Date'})
display(df.head())
```

```python
df['hour'] = df['Booking_Date'].dt.hour
df['weekday'] = df['Booking_Date'].dt.day_name()
df['month'] = df['Booking_Date'].dt.month
```

```python
df.head()
```

```python
df["fare"]=df["Ride Distance"]*2.5
```

```python
df.drop_duplicates(inplace=True)
```

```python
total_rides = len(df)

completed_rides = len(df[df['Booking Status'] == 'Completed'])
cancelled_rides = len(df[df['Booking Status'] == 'Cancelled'])

completion_rate = completed_rides / total_rides * 100
cancellation_rate = cancelled_rides / total_rides * 100

total_revenue = df[df['Booking Status'] == 'Completed']['fare'].sum()
average_fare = df['fare'].mean()
df.head()
```

```python
rides_per_day = df.groupby('weekday').size()

plt.figure(figsize=(8,5))
rides_per_day.plot(kind='bar')
plt.title("Rides per Weekday")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['fare'], bins=30, kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
df.to_csv("/content/ncr_ride_bookings.csv",index=False)
```
